```python
from nsepy import get_history
from datetime import date
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from pandas.plotting import scatter_matrix
import seaborn as sns
from nsepy import get_history
from datetime import date
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from pandas.plotting import scatter_matrix
import seaborn as sns
```

```python
data1=get_history(symbol="HINDUNILVR", start=date(2022,3,1), end=date(2022,8,30))
data2=get_history(symbol="ITC", start=date(2022,3,1), end=date(2022,8,30))
data3=get_history(symbol="COLPAL", start=date(2022,3,1), end=date(2022,8,30))
```

```python
data3.head()
```

```python
data4=get_history(symbol="NIFTY", start=date(2022,3,1), end=date(2022,8,30), index=True)
data4['Symbol']="NIFTY"
data4=data4[['Symbol','Open', 'High', 'Low', 'Close', 'Volume', "Turnover"]]
```

```python
data5=get_history(symbol="NIFTY FMCG", start=date(2022,3,1), end=date(2022,8,30), index=True)
data5['Symbol']="NIFTY FMCG"
data5=data5[['Symbol','Open', 'High', 'Low', 'Close', 'Volume', "Turnover"]]
```

```python
# A quick check to find columns that contain outliers
fig = plt.figure(figsize = (15, 7.2))
ax = sns.boxplot(data = data1.iloc[:, 0:13], orient = 'h')
```

```python
fig = plt.figure(figsize = (15, 7.2))
ax = sns.boxplot(data = data2.iloc[:, 0:13], orient = 'h')
```

```python
fig = plt.figure(figsize = (15, 7.2))
ax = sns.boxplot(data = data3.iloc[:, 0:13], orient = 'h')
```

```python
fig = plt.figure(figsize = (15, 7.2))
ax = sns.boxplot(data = data4.iloc[:, 0:13], orient = 'h')
```

```python
fig = plt.figure(figsize = (15, 7.2))
ax = sns.boxplot(data = data5.iloc[:, 0:13], orient = 'h')
```

```python
df = pd.concat([data1, data2, data3,data4,data5], axis=0)
df.tail()
```

```python
df.to_csv('Stock_Cleaned_Data.csv')
```

```python
df = df.reset_index()
df = df[['Date', 'Close', 'Symbol']]
df.head()
```

```python
df_pivot = df.pivot('Date','Symbol','Close').reset_index()
df_pivot.head()
```

```python
df_pivot.tail()
```

```python
# A quick check to find columns that contain outliers
fig = plt.figure(figsize = (15, 7.2))
ax = sns.boxplot(data = df.iloc[:, 0:13], orient = 'h')
```

```python
corr_df = df_pivot.corr(method='pearson')
#reset 'Symbol' as index (rather than 0-X)
corr_df.head().reset_index()
#del corr_df.index.name
corr_df.head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(corr_df, annot=True, cmap='RdYlGn')
plt.figure()
```

```python
sns.pairplot(df_pivot, diag_kind="hist")
```

```python
from matplotlib import rcParams
plt.plot(df_pivot.Date, df_pivot.COLPAL, color = "ORANGE", label = "COLPAL")
plt.plot(df_pivot.Date, df_pivot.HINDUNILVR, label = "HINDUNILVR")
plt.plot(df_pivot.Date, df_pivot.ITC, color = "RED", label = "ITC")
plt.plot(df_pivot.Date, df_pivot.NIFTY, color = "GREEN", label = "NIFTY")
plt.xlabel("DATE")
plt.ylabel("CLOSING PRICE")
rcParams['figure.figsize'] = 10,8
plt.legend()
plt.grid("True")
plt.title("HUL, ITC, COLPAL and NIFTY")
```
